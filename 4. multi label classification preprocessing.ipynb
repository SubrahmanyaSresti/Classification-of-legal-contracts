```python
# Obligation is much larger than the other classes, so we can downsample it to create balance
# Creating 2 dataframes majority and minority
# The majority df has tag == obligation and the minority df has tag != obligation
majority_df = df[df['tag'] == "['obligation']"]
minority_df = df[df['tag'] != "['obligation']"]

print(majority_df.shape)
print(minority_df.shape)
```

```python
# Retrieving important words from the dataset
comment_words = ' '

for val in df_new.sentence:
        bag = [ ]
        tokens = word_tokenize(val)
        tokens = [token.lower() for token in tokens if token.isalpha()]
        pos_tags = nltk.pos_tag(tokens)

        # Include modal verbs and other verb types
        select_pos = ['MD', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
        for word, tag in pos_tags:
            if tag in select_pos:
                bag.append((word))

        for i in range(len(bag)):
            bag[i] = bag[i].lower()
        comment_words += " ".join(bag)+" "
```

```python
# Creating word cloud using the 'important words' retrieved in the previous steps
word_cloud = WordCloud(
                    width = 800, height = 800,
                    background_color='white',
                    min_font_size = 10
                    ).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor = None)
plt.title("Word Cloud of Modal Words & Verbs", fontsize=20)
plt.imshow(word_cloud)
plt.axis("off")
plt.tight_layout(pad = 0)
```

```python
# Pre-processing the text
# Displaying nltk stopwords
stop_words = set(stopwords.words('english'))
print(f"Total stopwords: {len(stop_words)}")
print(stop_words)
```

```python
# Creating variable for nltk lemmatizer method
wnl = WordNetLemmatizer()

# Creating function to clean text
def clean_text(sentence):
     a = []
     
     # Splitting sentence into tokens
     tokens = word_tokenize(sentence)

     # Removig tokens that are not alphabetic (e.g. standalone punctuation)
     tokens = [token.lower() for token in tokens if token.isalpha()]

     # Reducing each word to its lemma
     for token in tokens:
          lem_word = wnl.lemmatize(token)
          a.append(lem_word)
     
     sentence = " ".join(a)
     return sentence

# Applying function to dataframe
df_new['clean_sentence'] = df_new['sentence'].apply(clean_text)
```
